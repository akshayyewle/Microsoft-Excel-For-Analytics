# **Setup**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline  
import datetime as dt
import yfinance as yf
```

# **Data Import**

```python
from numpy import repeat


Stock_Ticker = 'AAPL'
Start_Date = '1984-01-01'
End_Data = dt.date.today()

# Apple Share Price Data
Stock_Data = yf.download(Stock_Ticker, start=Start_Date, end=End_Data)

print('Dataframe Shape: ',Stock_Data.shape)
print('\n',('='*100))
Stock_Data.head(10)
```

## **Stock Price Plotting**

```python
plt.figure(figsize=(8,4))
plt.plot(Stock_Data.index,Stock_Data['Close'])

# iPod Launch
plt.axvline(pd.Timestamp('2001-10-23'), color="red",linestyle="--", label="iPod")
plt.text(pd.Timestamp('2001-10-23'),y=125,s='iPod Launch', color='r',rotation='vertical')

# iPhone Launch
plt.axvline(pd.Timestamp('2007-01-09'),color="red",linestyle="--", label="iPhone")
plt.text(pd.Timestamp('2007-01-09'),y=125,s='iPhone Launch', color='r',rotation='vertical')

# iPad Launch
plt.axvline(pd.Timestamp('2010-01-27'), color="red",linestyle="--", label="iPad")
plt.text(pd.Timestamp('2010-01-27'),y=125,s='iPad Launch', color='r',rotation='vertical')

# Apple Watch Launch
plt.axvline(pd.Timestamp('2015-04-24'), color="red",linestyle="--", label="iPad")
plt.text(pd.Timestamp('2015-04-24'),y=125,s='Apple Watch', color='r',rotation='vertical')

# Airpod Launch
plt.axvline(pd.Timestamp('2016-12-13'), color="red",linestyle="--", label="iPad")
plt.text(pd.Timestamp('2016-12-13'),y=125,s='Airpods', color='r',rotation='vertical')

# M1 Macbook Air Launch
plt.axvline(pd.Timestamp('2020-11-10'), color="red",linestyle="--", label="iPad")
plt.text(pd.Timestamp('2020-11-10'),y=125,s='M1 Macbook Air', color='r',rotation='vertical')

plt.title('Apple Share Price & Major Product Launches') 
plt.xlabel('Date')
plt.ylabel('Share Price (USD)')
plt.grid()
plt.show()
```
